```python
from googleads.adwords import AdWordsClient, ReportQueryBuilder
import pandas as pd
from io import StringIO
```

Переменные, передаваемые на вход

```python
client = AdWordsClient.LoadFromStorage('D:\\Python\\ads_statistics_scripts\\googleads.yaml')
```

```python
report_downloader = client.GetReportDownloader(version='v201809')
```

```python
report_query = (ReportQueryBuilder()
                  .Select('Id', 'BusinessName', 'AdType', 'CreativeFinalUrls')
                  .From('AD_PERFORMANCE_REPORT')
                #   .Where('Status').In('ENABLED', 'PAUSED')
                  .During('20200101, 20211110')
                  .Build())
```

```python
report_query = (ReportQueryBuilder()
                  .Select('CampaignId', 'CampaignName', 'CampaignTrialType', 'BiddingStrategyType', 'AdvertisingChannelType', 'AdvertisingChannelSubType', 'CampaignStatus', 'Cost', 'Clicks', 'Date')
                  .From('CAMPAIGN_PERFORMANCE_REPORT')
                  .During('20210101, 20211101')
                  .Build())
```

```python
report_query = (ReportQueryBuilder()
                  .Select('Id', 'Criteria', 'CriteriaType', 'CampaignId', 'FinalUrls')
                  .From('CRITERIA_PERFORMANCE_REPORT')
                #   .Where('IsNegative').In(True, False)
                  .During('20200101, 20211101') 
                  .Build())
```

```python
report_query = (ReportQueryBuilder()
                  .Select('Id', 'Criteria', 'CriteriaType', 'CampaignId')
                  .From('CAMPAIGN_CRITERIA_REPORT')
                  .Where('IsNegative').In(True, False)
                #   .During('') 
                  .Build())
```

```python
report_query = (ReportQueryBuilder()
                  .Select('BudgetId', 'BudgetName', 'TotalAmount')
                  .From('BUDGET_PERFORMANCE_REPORT')
                  # .Where('IsNegative').In(True, False)
                  # .During('20210101, 20211101') 
                  .Build())
```

```python
report_csv = report_downloader.DownloadReportAsStringWithAwql(
    report_query, 'CSV', skip_report_header=not False,
    skip_column_header=False, skip_report_summary=True,
    # include_zero_impressions=False
)
```

```python
df = pd.read_csv(StringIO(report_csv))
df['Cost'] = df['Cost'] / 1000000
df
```
